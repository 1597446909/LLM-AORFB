````python
id_container = ""
noinfo = []
for idx,fp in enumerate(files_path):
	# file_name = fp.split("\\")[-1].split("：")[1]
	file_name = fp.split("\\")[-1]
	print(file_name)
	# num = fp.split("\\")[-1].split("：")[0].replace("M","")
	file_object = client.files.create(file=Path(fp), purpose="file-extract")
	file_id = file_object.id
	# if idx+1 < len(files_path):
	# 	next_file = files_path[idx+1].split("\\")[-1]
	# 	if next_file.startswith(num+"S"):
	# 		id_container+=f'fileid://{file_id}'
	# 		print(idx,file_name)
	# 		continue
	# 	elif len(id_container)==0:
	# 		print(idx,file_name)
	# if id_container:
	# 	id_container+=f',fileid://{file_id}'
	# else:
	# 	id_container+=f'fileid://{file_id}'
	id_container=f'fileid://{file_id}'
	count = 0
	while True:
		try:
			completion = client.chat.completions.create(
			model="qwen-long", 
			messages=[
				{'role': 'system', 'content': 'You are a helpful assistant.'},
				{'role': 'system', 'content': id_container},
				{'role': 'user', 'content': abstract_prompt}
			],
			stream=False,
		)
			if completion.choices[0].message.content.replace("```","").replace("json","").replace(r'"',"").replace(r"'","").strip()=="None":
				noinfo.append(file_name)
				print(idx,"None\n\n")
				break
            
			output = re.findall(pattern,completion.choices[0].message.content)[0]
			print(idx,output,"\n\n")
			try:
				output = [json.loads(output)]
			except:
				output = re.findall(r'\{\s*"modification_site"\s*:\s*\d+,\s*"functional_group"\s*:\s*"[^"]*",\s*"application"\s*:\s*"[^"]*",\s*"report_content"\s*:\s*"[^"]*",\s*"feasibility_rating"\s*:\s*\d+\s*(?:,\s*)?\}',output)[0]
				if type(output)==str:
					output = [output]
			break
		except Exception as e:
			count+=1
			print(f"重试第{count}次",e,"\n\n")
			if count==5:
				print(file_name)
				break
	if completion.choices[0].message.content.replace("```","").replace("json","").replace(r'"',"").replace(r"'","").strip()=="None":
		continue
	if count==5:
		break
	if type(output)==list and type(output[0])==list:
		output=output[0]
	for out in output:
		if '$schema' in out:
				continue        
		if type(out)==str:
			out = json.loads(out)
		content = out["report_content"]
		embcompletion = client.embeddings.create(
			model="text-embedding-v4",
			input=content,
			dimensions=1024, # 指定向量维度（仅 text-embedding-v3及 text-embedding-v4支持该参数）
			encoding_format="float"
		)
		embedding = json.loads(embcompletion.model_dump_json())["data"][0]["embedding"]
		if file_id not in db:
			db[file_id] = {
				"embedding":[embedding],
				"report":[content],
				"file_name":file_name,
				"modification_site":[out["modification_site"]],
				"functional_group":[out["functional_group"]],
				"application":[out["application"]],
				"feasibility_rating":[out["feasibility_rating"]]}
		else:
			db[file_id]["embedding"].append(embedding)
			db[file_id]["report"].append(content)
			db[file_id]["modification_site"].append(out["modification_site"])
			db[file_id]["functional_group"].append(out["functional_group"])
			db[file_id]["application"].append(out["application"])
			db[file_id]["feasibility_rating"].append(out["feasibility_rating"])
	id_container = ""
````
